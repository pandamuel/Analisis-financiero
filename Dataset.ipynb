```python
ARMAS=filtro['ARMAS_MEDIOS'].unique()
ARMAS
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.bar(ARMAS, filtro['ARMAS_MEDIOS'].value_counts(),color='g')
ax.set_title('ARMAS')
ax.set_ylabel('Cantidad de personas')
ax.set_xlabel('ARMAS')
ax.set_xticklabels(ARMAS, rotation=90)
```

```python
count_ventana = filtro.rolling(window=2).count()
x_new= count_ventana .index
y_new= count_ventana .sum(axis='columns')
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(16, 8))
ax.plot(x_new, y_new, label='SEXO',color='r',marker='*')
ax.plot(x_new, y_new-300, label='Pr',color='b',marker='s')
ax.set_xlabel('Año')                 
ax.set_ylabel('Precipitación acumulada (mm.)')
ax.set_title('Precipitaciones acumuladas según el año')
ax.legend(loc='upper left')
```
